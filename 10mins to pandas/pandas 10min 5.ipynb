# Pandas Histogram

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

* 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처\
import matplotlib as mpl\
mpl.rcParams['axes.unicode_minus'] = False

```python
df4 = pd.DataFrame(
    {"a": np.random.randn(1000) + 1,
     "b": np.random.randn(1000),
     "c": np.random.randn(1000) - 1,},
    columns=["a", "b", "c"], )
df4.head()
```

```python
df4.plot.hist(alpha=0.5) # alpha : 투명도
```

```python
df4.plot.hist(stacked=True, bins=20) # bins : 막대개수, bin따라 freq도 바뀜
```

```python
df4["a"].plot.hist(orientation="horizontal", cumulative=True)
```

```python
df4["a"].diff().hist()
df4.diff().hist()
```

# Pandas Hexbin Plot

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=["a", "b"])
df["b"] = df["b"] + np.arange(1000) # 각 index값을 추가
```

```python
df.head()
df.tail()
```

```python
np.arange(1000) # 0부터 1000까지 숫자
```

```python
df.plot.hexbin(x="a", y="b", gridsize=25)
```

```python
df.plot.hexbin(x="a", y="b", gridsize=10)
```

```python
df["z"] = np.random.uniform(0, 3, 1000)
df.head()
```

```python
df.plot.hexbin(x="a", y="b", C="z", reduce_C_function=np.max, gridsize=25)
```

* reduce_C_function
기본적으로 각 (x,y) 주변의 개수에 대한 히스토그램이 계산됨.\
C 및 reduce_C_function 인수에 값을 전달하여 대체 집계를 지정 가능.\
C :  각 (x,y)점에서 값을 지정\
reduce_C_function : 빈의 모든값을 단일 숫자로 줄이는 인자.

```python
df.plot.hexbin(x="a", y="b", C="z", reduce_C_function=np.mean, gridsize=100)
```
